```
import numpy as np
import pandas as pd
import os
import urllib.request
from pandas_profiling import ProfileReport
```

```
def predict_encoding(file_path, n_lines=20):
    '''Predict a file's encoding using chardet'''
    import chardet
    with open(file_path, 'rb') as f:
        rawdata = b''.join([f.readline() for _ in range(n_lines)])

    return chardet.detect(rawdata)['encoding']
```

```
#Fone: Marinha do Brasil - Pesquisa feita em 2019
remote_file_name = "https://www.marinha.mil.br/dpc/sites/www.marinha.mil.br.dpc/files/dados-abertos/total_tipo_emb_2019.csv"
local_file_name = os.path.join("total_tipo_emb_2019.csv")
urllib.request.urlretrieve(remote_file_name, local_file_name) 
```

```
df = pd.read_csv("total_tipo_emb_2019.csv", encoding=predict_encoding("total_tipo_emb_2019.csv") )
```
